### What is Linear Regression?

Linear regression is the predictive model used to find the relationship between dependent variable or one or more independent variable.

Linear: means straight line
Regression: Predicting the contineous values or real numbers.

It is a type of regression machine learning which is the comes under the supervised machine learning algorithm that computes the linear relationship between a dependent variable and one or more independent features. **When the number of the independent feature, is 1 then it is known as Univariate Linear regression, and in the case of more than one feature, it is known as multivariate linear regression.**

### Why Linear Regression is Important?

Whenever we have to predict the dependent variable (Y) on the basis independent variables(X) then linear regression plays a very important role.

**IMP_Note: Inshort with the help of Linear regression, we can pick the points of which SSR(sum of square of residual) is less and calculated the relationship between dependent(Y) and independent variables(X)**

**Practically:** Linear regression helps us to find the m(slope) and c(intercept) values from y=mx+c by taking the least number of SSR so that we can predict the output with more accuracy.

### Types of Linear Regression:

**1) Simple Linear Regression:**


When we are predicting the dependent variable (Y) on the basis of only one independent variable (X) then this type of linear regression is known as simple linear regression.


formula: y=mx+c

*m=slope*
  
*c=intercept*

*x=independent variable*

*y=dependent variable*

**2) Multiple Linear Regression:**

When we are predicting the dependent variable (Y) on the basis two or more independent variable (X) then this is known as Multiple Linear Regression.

*formula:  y = m1x1 + m2x2 + m3x3 + m4x4 + .... + c*

*m1,m2,m3,m4 - slopes*
  
*x1,x2,x3,x4 -  independent variables*
  
*Y - Depedent variable*

*c - Intercept*

### Assumptions in Linear Regression:

In linear regression, assumptions refer to the underlying conditions that need to be satisfied for the model to provide reliable and meaningful results. 

```python
1) Linearity 
2) No Multicolinearity
3) Normality of residual
4) Homosceasticity (Residual)
```

#### Linearity:

### Gradient Descent algorithm

**What is Gradient Descent algorithm?**

Ans:It is an iterative algorithm through which we can find the minimum value of the error point that error point are also known as cost function for linear regression.

Procedure or working of this Gradient descent algorithm:

1) This algorithm or model will pick any two random values of slope(m) & intercept(c) to calculate the new predicted slope(m) and intercept(c) values.

2) It uses partial derivates to find the new slope(m) and intercept(c) values.

3) We can get the best m&c values from the infinite number of possibilities.

4) Step size becomes smaller and smaller till we reach the global minima points and that global minima point values are best slope and intercept values also known as least Mean Sqaure Error (MSE) value.

MSE(Mean Sqaure Error)=SSE/N , where N is number of samples or number of error points
